```python
sys.path.insert(0, './common')

import sys
import cartpole
import qlearning_target_model
```

```python
env = cartpole.CartPoleEnv()
```

```python
hidden_layer = 128
model = qlearning_target_model.QLearning(4, 2, hidden_layer, lr=1e-4, mem_size=10_000)
```

```python
model.load_model()
```

```python
model.set_train_params(plot_smooth=10, gamma=0.99, batch_size=256, sync_interval=512, epsilon_decay=0.995, learn_interval=5)
model.train(env, 1250, show_plot=True)
```

```python
model.save_model()
```

```python
model.plot_train_memory(smooth=100)
```

```python
model.test(env)
```

```python
model.test_avg(env, 10)
```
